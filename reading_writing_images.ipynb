```python
import rasterio
from rasterio.plot import show
import numpy as np

# define file path
fp = r"data/hitscor_sonneberg_.tif"

# create reader
raster = rasterio.open(fp,mode="r+")

# plot raster
show(raster)

# Printing Raster Information
print("Type: ",type(raster))
print("CRS: ",raster.crs)
print("Width: ",raster.width)
print("Height: ",raster.height)
print("Bands: ",raster.count)
print("BBox: ",raster.bounds)

# extract one band as np.array
band1 = raster.read(1)
# manipulate band so that it is identifieable
band2 = np.where(band1==0, 1, band1)
# write bew band over old band
raster.write (band2, indexes = 40)

# close reader
raster.close()
```
